## Get domain shapes here ($\mathfrak{D}_{\mathrm{eff}}$)

```python
import numpy as np
import matplotlib.pyplot as py
```

Change `V0` and `m` to match the parameters of interest

```python
V0 = 1 #Units of 10^-10 Mpl
m = 0.1

philine = np.linspace(0, 16, 101)

py.plot(philine, [np.sqrt(0.5*V0*(p**m)) for p in philine], label = '$ε = 1$')
py.plot(philine, [np.sqrt((1/3)*V0*(p**m)) for p in philine], label = '$ε = 0$')
py.fill_between(philine, [np.sqrt((1/3)*V0*(p**m)) for p in philine], [np.sqrt(0.5*V0*(p**m)) for p in philine], alpha = 0.1)
py.xlabel('$\phi/ \, M_{pl}$'), py.ylabel('$H/ \, 10^{-5}M_{pl}$')
py.legend()

py.figure()
```
